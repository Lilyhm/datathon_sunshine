## Distribution of Activities

```python
activity_counts = internal.pivot_table(values='Hub_ActivityID', index='HubRandomID', aggfunc=len)['Hub_ActivityID']

activity_counts.plot.kde()
plt.xlim(activity_counts.min(), activity_counts.max())

plt.axvline(activity_counts.mean(), color = 'Black', label = 'Median')
plt.axvline(activity_counts.median(), color = 'Green', label = 'Mean')
plt.axvline(activity_counts.mode()[0], color = 'Orange', label = 'Mode')

```
